```python
# Use bisection to determine an interval of length less than 1/8 containing 
# a solution to x^5 + x − 20 = 0, starting with the interval [1, 2].

fline = lambda x: x ** 5 + x  - 20

def bisection_method(f, a, b, tol):
    if (f(a) * f(b)) > 0:
        raise Exception('There is no root at this bound')

    error = (b - a)
    mid = (a + b) / 2

    if tol + 1e-9 > error:
        print(f'With tolerance of {tol}, the root is: {mid}')
    elif (f(a) * f(mid)) <= 0:
        return bisection_method(f, a, mid, tol)
    else:
        return bisection_method(f, mid, b, tol)

bisection_method(fline, 1, 2, 1/8)
```

```python
# Suppose we approximate √3 by using bisection to find the root of the equation
# f(x) = 0, where f(x) = x^2 − 3.

# If we choose a starting interval whose endpoints are 0 and 2, how many iterations
# (ie. how many midpoint approximations) are required in order that the
# absolute error in our approximation of the root is no greater than 0.0001

fline = lambda x: x ** 2 - 3

iterations = 0

def bisection_method(f, a, b, tol):
    global iterations

    if (f(a) * f(b)) > 0:
        raise Exception('There is no root at this bound')

    error = (b - a)
    mid = (a + b) / 2
    iterations += 1

    if tol + 1e-9 > error:
        print(f'With tolerance of {tol}, the root is: {mid}, running {iterations} iterations')
    elif (f(a) * f(mid)) <= 0:
        return bisection_method(f, a, mid, tol)
    else:
        return bisection_method(f, mid, b, tol)

bisection_method(fline, 0, 2,  0.0001)
```
