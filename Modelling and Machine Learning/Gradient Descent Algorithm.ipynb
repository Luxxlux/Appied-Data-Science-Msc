### Useful formulae

Given a set of $P$ points $\{ (x_i, y_i)\}$, with $i = 1, \dots, P$, the mean squared error cost function 
$g_{\mathrm{MSE}}$ for the linear interpolation $y = a x + b$ is given by 

\begin{equation*}
g_{\mathrm{MSE}} = \frac{1}{P} \sum_{i = 1}^P \left(a x_i + b - y_i \right)^2 \ .
\end{equation*}

Its gradient is given by

\begin{equation*}
\vec{\nabla} g_{\mathrm{MSE}} = 
\left(
\begin{array}{c}
\frac{\partial g_{\mathrm{MSE}}}{\partial a}\\
\frac{\partial g_{\mathrm{MSE}}}{\partial b}
\end{array}
\right)
= 
\left(
\begin{array}{c}
\frac{2}{P} \sum_{i = 1}^P x_i \left(a x_i + b - y_i \right)\\
\frac{2}{P} \sum_{i = 1}^P \left(a x_i + b - y_i \right)
\end{array}
\right) \ .
\end{equation*}


### Exercise - Linear regression and gradient descent

Using the mean squared error as the cost function, determine
analitically the parameters $a$ and $b$ of the linear 
regression $f(x) = a x + b$ applied to the points

\begin{align*}
(3.55, 7.53) \ ,\ &
(2.38, 3.67) \ , \

\end{align*}

Then, using the gradient descent with starting points $a_0 = 1$, $b_0 = 0$, learning rate $\alpha = 0.02$ and 1000 iterations (or epochs), determine numerically $a$ and $b$. 

### Model solution

We start by defining the inputs, with the arrays $x$ and $y$
holding the data and numdata containing the number of data. 

```python
import numpy as np
numdata = 8

x = np.array([3.55, 2.38, 4.10, 2.81, 1.05, 1.87, 1.40, 4.89])
y = np.array([7.53, 3.67, 8.21, 5.88, 1.6, 4.24, 2.82, 10.30])
```

Then, we move to the numerical calculation. Defining the single contribution of the point $p$ to the cost function $g$. Below, $para = a$, $parb = b$, $bindex = 0, 1$ is the index of the minibatch number and bsize is the batch size.

We also define the gradient $ggp$, which is the gradient evaluated at $\vec{x}_p$, as well as the total cost function $g$ (called gcost)

```python
P = numdata #Number of data points

def ggp (para1, parb1):
    gvalue = np.array([0.,0.])
    
    for i in range(P):
        fact = (para1*x[i] + parb1 - y[i])
        gvalue[0] += fact*x[i]
        gvalue[1] += fact
    
    return 2*gvalue/P

def gcost(para1, parb1):
    gacc = 0.0
    
    for i in range(P):
        gacc +=  (para1*x[i] + parb1 - y[i])**2
        
    return gacc/P
        
```

We define the number of epochs nepochs and the initial point $(a_0,b_0) \equiv \mathrm{(para,parb)}$

```python
nepochs = 1000
parain = 1.0
parbin = 0.0
```

```python
para = parain
parb = parbin
alpha = 0.02

for j in range(nepochs):    
    grad = ggp (para, parb)
    para -= alpha*grad[0]
    parb -= alpha*grad[1] 
        
    print ("Epoch:", j, "\t a = %2.6f"%(para), "\t b = %2.6f"%(parb))
```

Finally, we compue the difference of the cost function at the beginning and at the end:

```python
costdiff = gcost(parain,parbin) - gcost(para,parb)
print(gcost(parain,parbin),gcost(para,parb))
print("Difference in cost: %2.6f"%(costdiff))
```

Then, using the gradient descent with starting points $a_0 = 1$, $b_0 = 0$, learning rate $\alpha = 0.02$ and 1000 iterations (or epochs), determine numerically $a$ and $b$. 

```python
import numpy as np
x = np.array([3.70, 0.49, 3.84, 3.32, 3.83, 1.39, 0.60, 2.10])
y = np.array([-0.33, 2.52, -2.01, 0.26, -0.51, 1.94, 2.47, 0.82])
```

```python
def dgmse(a,b,x,y):
    num = len(x)
    running_total_a = 0
    running_total_b = 0

    for i in range(num):
        running_total_a += (a*x[i] + b - y[i]) * x[i]
        running_total_b += (a*x[i] + b - y[i])

    return (2*running_total_a)/num, (2*running_total_b)/num

dgmse(1,2,x,y)
```

```python
def gmse(a, b, x, y):
    num = len(x)
    running_total = 0

    for i in range(num):
        running_total += (a*x[i] + b - y[i])**2

    return  (running_total)/num

gmse(1,2,x,y)
```

```python
epochs = 1000
a_start = 1
b_start = 0
learning_rate = 0.02

updated_a = a_start
updated_b = b_start

for i in range(epochs):
    grad_a, grad_b = dgmse(updated_a, updated_b, x, y)

    updated_a = updated_a - (grad_a * learning_rate)
    updated_b = updated_b - (grad_b * learning_rate)

    error = gmse(updated_a, updated_b, x, y)

    print('Iteration:', i, "\t a: ", np.round(updated_a, 5), '\t b:', np.round(updated_b, 5), '\t Mean squared error:', np.round(error, 5))

```
