Linregress function

Instead of all these calculations, we can also use linregress class from scipy-stats module:

from scipy stats import linregress linregress (x, y)

```python
from scipy.stats import linregress
```

```python
lr = linregress(lenght, birthweight)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xvalues = np.linspace(min(lenght), max(lenght), 100)
yvalues = lr.intercept + lr.slope * xvalues
plt.scatter(lenght, birthweight)
plt.plot(xvalues, yvalues, color = 'r')
plt.show()
```

```python
lr.rvalue
# Moderate correlation
```

```python
df1 = pd.read_csv('Experience-Salary.csv')
df1
```

```python
import matplotlib.pyplot as plt
import numpy as np

exp_months = df1['exp(in months)']
sal_thou = df1['salary(in thousands)']


lr2 = linregress(exp_months, sal_thou)


xvalues = np.linspace(min(exp_months), max(sal_thou), 100)
yvalues = lr2.intercept + lr2.slope * xvalues
plt.scatter(exp_months, sal_thou)
plt.plot(xvalues, yvalues, color = 'r')
plt.show()
```
