```python
from scipy.stats import uniform
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2, 9, 1000)
plt.plot(x, uniform.pdf(x, loc = 1, scale = 5))
plt.show()
```

## Cumulative distribution function

```python
x = np.linspace(-2, 9, 1000)
plt.plot(x, uniform.cdf(x, loc=1, scale=5))
plt.show()
```

## Generating random values

```python
import pandas as pd
x = uniform.rvs(size = 10**6, loc = 0, scale = 1, random_state = 123)
x.mean()
```

## 2. Exponential distribution

```python
from scipy.stats import expon
expon.pdf(1, scale=5)
```

```python
x = expon.rvs(size = 10**6, scale = 5, random_state = 12)
x.var()
```

## Percentiles

```python
a = expon.ppf(0.3, scale = 1/0.2)
a
```

```python
expon.cdf(a, scale = 5)
```

```python
b = 0.4 + expon.cdf(1, scale=1/0.7)
expon.ppf(b, scale=1/0.7)
```

## 3. Normal Distribution

```python
from scipy.stats import norm
```

```python
x = np.linspace(-10, 14, 1000)
y = norm.pdf(x, loc = 2, scale = 3)
plt.plot(x, y, color = 'red')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```

```python
x = np.linspace(-10, 14, 1000)
y = norm.cdf(x, loc = 2, scale = 3)
plt.plot(x, y, color = 'green')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```
