Define the function and we expand the expression at the same step

```python
f = expand((x-1)*(x-2)*(x-3))
f
```

Next, we find the derivative of F, assign it to the new function F1, and then we solve the critical points equation

```python
f_1 = diff(f, x)
f_1
```

The following code will solve to find the roots of the polynomial

```python
sol_list = solve(f_1, x)
print(f'Root one:')
sol_list[0]
```

```python
print(f'Root two:')
sol_list[1]
```

```python
type(sol_list)
```

Now we assign the first element of the list (with index 0) to x1, and the second to x2

```python
x_1 = sol_list[0]
x_2 = sol_list[1]
```

```python
# This puts the value in a decimal
(x_1).evalf()
```

```python
(x_2).evalf()
```

Next we find the second derivative of f, which we compute as derivative of f1, and assign it to the function f2

```python
f_2 = diff(f_1, x)
f_2
```

Then, we compute f2(x1) and f2(x2) by using 'subs' introduction

```python
# Substitutes with the x symbols, the x1 funciton within f2
f_2.subs(x,x_1)
```

```python
f_2.subs(x,x_2)
```

We conclude taht x1 is a local maximum and x2 is a local minimum of f

Next we find the minimum value of f(x1) and maximal value of f(x2), using 'subs' and asking sympy to simplify the output

```python
simplify(f.subs(x,x_1))
```

```python
simplify(f.subs(x,x_2))
```

We can also check numerical value using 'evalf()', for instance we find that f(x2) is roughly -0.38 as follows

```python
simplify(f.subs(x,x_2)).evalf()
```

```python
simplify(f.subs(x,x_1)).evalf()
```
